#creating an empty figure with 2 axes

```python
fig1, (ax,ax1) = plt.subplots(1,2)
ax.barh(0,0)
ax1.barh(0,0)
plt.show()
```

```python
fig1, (ax,ax1) = plt.subplots(1,2)
ax.barh(0,0)
ax1.barh(0,0)
ax.set_title('Age Histogram')
ax1.set_title('Politcal Identity')
plt.show()
```
